```python
df_PG=df[df['Position']=='PG']
df_SF=df[df['Position']=='SF']
df_PF=df[df['Position']=='PF']
df_SG=df[df['Position']=='SG']
df_C=df[df['Position']=='C']
```

```python
mean_pg_age,mean_pg_wt=np.mean(df_PG['Age']),np.mean(df_PG['Weight'])
mean_sf_age,mean_sf_wt=np.mean(df_SF['Age']),np.mean(df_PG['Weight'])
mean_pf_age,mean_pf_wt=np.mean(df_PF['Age']),np.mean(df_PG['Weight'])
mean_sg_age,mean_sg_wt=np.mean(df_SG['Age']),np.mean(df_PG['Weight'])
mean_c_age,mean_c_wt=np.mean(df_C['Age']),np.mean(df_PG['Weight'])
```

```python
median_pg_age,median_pg_wt=np.median(df_PG['Age']),np.median(df_PG['Weight'])
median_sf_age,median_sf_wt=np.median(df_SF['Age']),np.median(df_PG['Weight'])
median_pf_age,median_pf_wt=np.median(df_PF['Age']),np.median(df_PG['Weight'])
median_sg_age,median_sg_wt=np.median(df_SG['Age']),np.median(df_PG['Weight'])
median_c_age,median_c_wt=np.median(df_C['Age']),np.median(df_PG['Weight'])
```

```python
min_pg_age,min_pg_wt=np.min(df_PG['Age']),np.min(df_PG['Weight'])
min_sf_age,min_sf_wt=np.min(df_SF['Age']),np.min(df_PG['Weight'])
min_pf_age,min_pf_wt=np.min(df_PF['Age']),np.min(df_PG['Weight'])
min_sg_age,min_sg_wt=np.min(df_SG['Age']),np.min(df_PG['Weight'])
min_c_age,min_c_wt=np.min(df_C['Age']),np.min(df_PG['Weight'])
```

```python
max_pg_age,max_pg_wt=np.max(df_PG['Age']),np.max(df_PG['Weight'])
max_sf_age,max_sf_wt=np.max(df_SF['Age']),np.max(df_PG['Weight'])
max_pf_age,max_pf_wt=np.max(df_PF['Age']),np.max(df_PG['Weight'])
max_sg_age,max_sg_wt=np.max(df_SG['Age']),np.max(df_PG['Weight'])
max_c_age,max_c_wt=np.max(df_C['Age']),np.max(df_PG['Weight'])
```

```python
std_pg_age,std_pg_wt=np.std(df_PG['Age']),np.std(df_PG['Weight'])
std_sf_age,std_sf_wt=np.std(df_SF['Age']),np.std(df_PG['Weight'])
std_pf_age,std_pf_wt=np.std(df_PF['Age']),np.std(df_PG['Weight'])
std_sg_age,std_sg_wt=np.std(df_SG['Age']),np.std(df_PG['Weight'])
std_c_age,std_c_wt=np.std(df_C['Age']),np.std(df_PG['Weight'])
```
