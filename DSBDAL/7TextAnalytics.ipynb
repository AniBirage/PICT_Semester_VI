```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
import re
import math
```

```python
with open("TextFile.txt", "r") as file:
    text = file.read()
text
```

```python
tokens = word_tokenize(text)
tokens = [token for token in tokens if re.match("^[a-zA-Z0-9]+$", token)]
print(tokens)
```

```python
pos_tags = pos_tag(tokens)
for word, pos_tag in pos_tags:
    print(word, ":", pos_tag)
```

```python
stop_words = set(stopwords.words("english"))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print(filtered_tokens)
```

```python
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
print(stemmed_tokens)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print(lemmatized_tokens)
```

```python
total_frequency = len(lemmatized_tokens)

tf_dict = {}
for word in lemmatized_tokens:
    if word in tf_dict:
        tf_dict[word] += 1
    else:
        tf_dict[word] = 1

for key, value in tf_dict.items():
    value = float(value) / float(total_frequency)
    tf_dict[key] = value

for word, frequency in tf_dict.items():
    print(word, ":", frequency)
```

```python
idf_dict = {}
N = 1

for word in tf_dict:
    doc_count = sum(1 for doc_tf in [tf_dict] if word in doc_tf)
    idf_value = math.log10(N / (1 + doc_count)) + 1
    idf_dict[word] = idf_value

for word, idf_value in idf_dict.items():
    print(word, ":", idf_value)
```
