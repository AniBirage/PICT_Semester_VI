```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import set_theme

titanic = sns.load_dataset('titanic')
titanic
```

```python
set_theme()

#Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='fare', data=titanic)
plt.title('Line Plot of Age vs Fare')
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='fare', data=titanic)
plt.title('Scatter Plot of Age vs Fare')
plt.show()
```

```python
#Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='age', data=titanic)
plt.title('Box Plot of Age by Class')
plt.show()
```

```python
#Point Plot
plt.figure(figsize=(10, 6))
sns.pointplot(x='class', y='age', data=titanic)
plt.title('Point Plot of Age by Class')
plt.show()
```

```python
#Count Plot
plt.figure(figsize=(10, 6))
sns.countplot(x='class', data=titanic)
plt.title('Count Plot of Passenger Class')
plt.show()
```

```python
#Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='class', y='age', data=titanic)
plt.title('Violin Plot of Age by Class')
plt.show()
```

```python
#Swarm Plot
plt.figure(figsize=(10, 6))
sns.swarmplot(x='class', y='age', data=titanic)
plt.title('Swarm Plot of Age by Class')
plt.show()
```

```python
#Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='class', y='fare', data=titanic)
plt.title('Bar Plot of Fare by Class')
plt.show()
```

```python
#KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=titanic['age'], shade=True)
plt.title('Kernel Density Estimation (KDE) Plot of Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic['fare'], bins=30, kde=True, color='red')
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
